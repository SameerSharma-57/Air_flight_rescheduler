```python
from datetime import datetime

# Example date and time strings
date_string = '2023-12-09'
time_string = '15:35'

# Combine date and time strings and define the format
datetime_string = f'{date_string} {time_string}'
date_time_format = '%Y-%m-%d %H:%M'

# Parse the combined string to create a datetime object
date_time_object = datetime.strptime(datetime_string, date_time_format)

# Now, date_time_object contains the combined date and time
print(date_time_object)
```
